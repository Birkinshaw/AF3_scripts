```python
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem import AllChem
```

```python
m = Chem.MolFromSmiles('CCCCCCCC/C=C\\CCCCCCCC(=O)OC(COC(=O)CCCCCCCCCCCCCCC)COP(=O)([O-])OCC[N+](C)(C)C')
```

```python
#print(Chem.MolToMolBlock(m))
img = Draw.MolToImage(m)
```

```python
img
```

```python
m3 = Chem.AddHs(m)
```

```python
params = AllChem.ETKDGv3()
```

```python
AllChem.EmbedMolecule(m3, params)
```

```python
print(Chem.MolToMolBlock(m3),file=open('PC.mol','w+'))
```
